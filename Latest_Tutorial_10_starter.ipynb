```python
# load and read the quakers_nodelist.csv
with open('quakers_nodelist.csv','r') as nodecsv:
  nodereader = csv.reader(nodecsv)
  nodes = [n for n in nodereader][1:] # first line

print(nodes)
```

```python
# load and read the quakers_edgelist.csv
with open('quakers_edgelist.csv','r') as edgecsv:
  edgereader = csv.reader(edgecsv)
  edges = [n for n in edgereader][1:] # first line

print(edges)
```

```python
#create graph object
G = nx.Graph() # empty graph

#add lists of nodes and edges
nodenames = [n[0] for n in nodes] # extract the first value
G.add_nodes_from(nodenames)
```

```python
G.add_edges_from(edges)
```

```python
print(G.number_of_edges())
print(G.number_of_nodes())
```

NetworkX provides two convenient functions for adding attributes to all of a Graph’s nodes or edges at once:

**nx.set_node_attributes()** and **nx.set_edge_attributes()**.

To use these functions, you’ll need your attribute data to be in the form of a Python dictionary, in which node names are the keys and the attributes you want to add are the values.

```python
job_dict = {} # historical significance -> job
gender_dict = {}
birth_dict = {}
death_dict = {}
id_dict = {}

for node in nodes:
  job_dict[node[0]] = node[1]
  gender_dict[node[0]] = node[2]
  birth_dict[node[0]] = node[3]
  death_dict[node[0]] = node[4]
  id_dict[node[0]] = node[5]

print(job_dict)
```

```python
# Extract attributes from the quakers_nodelist.csv file and insert them into the dictionaries declared above
# Add each dictionary as a node attribute to the Graph object
nx.set_node_attributes(G, name='job', values=job_dict)
nx.set_node_attributes(G, name='gender', values=gender_dict)
nx.set_node_attributes(G, name='birthyear', values=birth_dict)
nx.set_node_attributes(G, name='deathyear', values=death_dict)
nx.set_node_attributes(G, name='ID', values=id_dict)

print(G.nodes['James Nayler'])
```

# NetworkX Functions

<i> Using built-in functions from NetworkX to calculate network metrics</i>

```python
# Calculate network density (how well the nodes are connected)
print("Network density", nx.density(G))
```

```python
# Identify shortest paths
# John Story Francis Bugg
path = nx.shortest_path(G, source="Francis Bugg", target="George Keith")
print("Shortest path between George and Francis ", path)
```

```python
# Identifying connected graph (i.e. graphs that only have one component) [Lecture ? : Components]
print(nx.is_connected(G))
# Next, use nx.connected_components to get the list of components,
# then use the max() command to find the largest one:
components = nx.connected_components(G)

# Create a "subgraph" of just the largest component
# Then calculate the diameter of the subgraph, just like you did with density.
largest_cmp = max(components, key=len)
print(len(largest_cmp))
```

```python
# Calculate node degree
deg_dict = dict(G.degree(G.nodes()))
nx.set_node_attributes(G, name='degree', values=deg_dict) # dict for the node degree
print(G.nodes['James Nayler'])
```

```python
# Sort the degrees of nodes
sorted_degree = sorted(deg_dict.items(), key=itemgetter(1), reverse=True) # highest to lowest
sorted_degree
```

```python
# Calculate node betweenness and eigenvector
bet_deg = nx.betweenness_centrality(G)
eig_deg = nx.eigenvector_centrality(G)

nx.set_node_attributes(G, name="betweenness", values=bet_deg)
nx.set_node_attributes(G, name="eigenvector", values=eig_deg)
```

```python
print(G.nodes['James Nayler'])
```

```python
# Sort nodes by key/values e.g. betweenness
sorted_bet = sorted(bet_deg.items(), key=itemgetter(1), reverse=True)
print("Top 10 nodes by betweenness centrality:")
for n in sorted_bet[:10]:
  print(n)
```

```python
#First get the top ??? nodes by betweenness as a list
top_bet = sorted_bet[:10]

#Then find and print their degree
for tb in top_bet:
  degree = deg_dict[tb[0]]
  print("Name:", tb[0], "| Betweenness:", tb[1], "| Degree:", degree)
```

```python
import community.community_louvain as cm
partitions = cm.best_partition(G)

plt.figure(figsize=(150,150))
nx.draw_networkx(G, pos=nx.spring_layout(G))
plt.show()
```

```python
print(partitions)
```

```python
nx.set_node_attributes(G, values = partitions, name = "modularity")
```

```python
modularity = {} # Create a new, empty dictionary
for k,v in partitions.items(): # Loop through the community dictionary
    if v not in modularity:
        modularity[v] = [k] # Add a new key for a modularity class the code hasn't seen before
    else:
        modularity[v].append(k) # Append a name to the list for a modularity class the code has already seen

for k,v in modularity.items(): # Loop through the new dictionary
    #if len(v) > 2: # Filter out modularity classes with 2 or fewer nodes
        print('Class '+str(k)+':', v) # Print out the classes and their members
```
